```python
np.unique(A_2, return_counts=True) #On observe deux tableaux, le premier contient les différentes entités du tableau et le deuxième renseigne sur le nombre d'occurence
```

```python
values,counts = np.unique(A_2, return_counts=True)
```

```python
counts.argsort()
```

```python
for i,j in zip(values[counts.argsort()],counts[counts.argsort()]):
    print(f'valeur {i} apparait {j} fois')
```

# NaN (Not a Number) Corrections 
Pour effectuer de l'algèbre linéaire avec Numpy on peut utiliser "linalg" (Voir documentation sur le site de Numpy)

```python
Z_2= np.random.randn(6,6)
Z_2
```

```python
Z_2[1:3,1:3]=np.nan
Z_2
```

```python
Z_2.mean()
```

```python
np.nanmean(Z_2), np.nanvar(Z_2),np.nanstd(Z_2)#Permet de calculer la moyenne, la variance ou l'écart-type tout en ignorant les NaN qu'on pourrait avoir dans notre tableau
```

##### Comment compter le nombre de NaN dans un tableau

```python
np.isnan(Z_2)#Cectte opération nous retourne un mask avec la variable "True" chaque fois qu'un NaN est une entité du tableau
```

```python
np.isnan(Z_2).sum() #Nous donne la somme des NaN dans le tableau
```

```python
np.isnan(Z_2).sum()/Z_2.size#Proportion des NaN dans le tableau
```

```python
Z_2[np.isnan(Z_2)]= 3
```

```python
Z_2
```
