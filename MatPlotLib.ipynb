### Cette bibbliothèque permet de tracer tous les graphiques qu'on pourrait imaginer, c'est une bibliothèque qui sert à la visualisation des données

1-Un graphique permet de mieux visualiser le problème sur lequel on travaille
2-Il sert à comprendre le problème, expliquer un phénomène
3-Pour créer les graphiques dans MatPlotlib il existe deux méthodes : Une méthode orientée objet (OOP) et une méthode plus basique (plt.plot)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(-10,30,2)
y= x**2
```

```python
plt.plot(x,y)
plt.show() #Uniquement si on utilise un interpréteur python autre que jupyter notebook
```

```python
x_1= np.arange(1,15,1)
y_1= np.arange(3,19,1)#RQ: dans matplotlib il faut que les deux tableau ont les memes nombres d'entités si non le programme retourne un message d'erreur.
                        #Pour cet exercice par exemple il faudrait modifier la valeur de 19 à 17 par exemple
plt.plot(x_1,y_1)
```

```python
x_2= np.arange(1,15,1)
y_2= np.arange(3,17,1)
plt.scatter(x_2,y_2) #La fonction scatter permet d'avoir un nuage de points
```

```python
plt.plot(x,y,c='red',lw = 3, ls = '--') #le paramètre c permet de choisir la couleur de la courbe, le paramètre lw permet de définir l'épaisseur
```

```python
plt.figure(figsize=(8,5))#Cette fonction permet de définir la taille de la figure en inch, pas en cm
plt.plot(x,y, label = 'courbe 1')
plt.plot(x,x, label = 'courbe 2')
plt.title('figure 1')
plt.xlabel('axe des abcisses')
plt.ylabel('axe des ordonnées')
plt.legend()
plt.show()
plt.savefig('reprise_avec_Matplotlib.png')

plt.figure(figsize=(8,5))#Pour créer une autre figure en dessous de la première, il suffit juste de faire un retour 
                        # à la ligne puis laisser un espace et écrire à nouveau "plt.figure"
plt.plot(x_2,y_2)
```

### La fonction Subplot()
Elle permet d'afficher plusieurs graphiques sur une seule et meme figure

```python
plt.subplot(2,2,1)#le paramètre 2 renvoi au nombre de lignes que l'on veut créer et 1 le nomvre de colonne et le dernier 1 la position
plt.plot(x,y, label = 'courbe 1')
plt.title('figure 1')
plt.xlabel('axe des abcisses')
plt.ylabel('axe des ordonnées')
plt.legend()
#plt.show()
plt.subplot(2,2,2)
plt.plot(x,y, label = 'courbe 3', c= 'red')
plt.title('figure 1')
plt.xlabel('axe des abcisses')
plt.ylabel('axe des ordonnées')
plt.legend()
#plt.show()
plt.subplot(2,2,3)
plt.plot(x,y, label = 'courbe 2', c= 'black')
plt.title('figure 1')
plt.xlabel('axe des abcisses')
plt.ylabel('axe des ordonnées')
plt.legend()
plt.show()
```

### La méthode orienté objet OOP

```python
fig,ax=plt.subplots(2,1, sharex=True) #L'avantage avec cette fonction est qu'on peut créer des graphes qui partagent le meme abcisse
ax[0].plot(y,np.cos(y))
ax[1].plot(x,np.cos(x))
plt.show()
```

## Graphique 3D avec Matplotlib

```python
#%matplotlib
from mpl_toolkits.mplot3d import axes3d
plt.figure(figsize=(12,8))
ax = plt.axes(projection = '3d')

ax.scatter(x,y,x**2, c=y)
```

```python
f = lambda x,y: np.cos(x)+np.sin(x+y)
```

```python
plt.figure(figsize=(12,8))
X=np.linspace(0,10,100)
Y=np.linspace(0,10,100)
X,Y = np.meshgrid(X,Y)
Z=f(X,Y)
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z, cmap='plasma')
```

```python
plt.hist(x, bins=100)
plt.hist(y, bins =50)
```

```python
plt.contour(X,Y,Z, 15, colors = 'blue')
```

```python
plt.contourf(X,Y,Z, 15, cmap = 'RdGy')
plt.colorbar()
```
